* date: Date of chart
* rank: Rank of song
* song: Song title

```python
# 定義函數以便執行肩並肩的列印
def side_by_side(*objs, **kwds):
    from pandas.io.formats.printing import adjoin
    space = kwds.get('space', 4)
    reprs = [repr(obj).split('\n') for obj in objs]
    print (adjoin(space, *reprs))
    print()
    return
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

import plotly.graph_objs as go
import plotly as py
from plotly.offline import init_notebook_mode, iplot

from math import pi
from pandas.plotting import parallel_coordinates

%matplotlib inline
```

### 資料集

```python
song_data = pd.read_csv('charts.csv')
```

```python
side_by_side(song_data.shape, song_data.isnull().sum(), 
            song_data.count())
song_data.head()
```

```python
song_data[song_data['last-week'].isnull()] #找尋last-week為空值
```

```python
song_data['last-week'] = song_data['last-week'].fillna(0) # last-week的NaN以0取代
print(song_data.isnull().sum()) #檢查是否沒有空值了
```

```python
# 將date提取出"年"與"月份"
from datetime import datetime
song_data['date'] = song_data['date'].astype('datetime64[ns]')
song_data['Year'] =song_data['date'].dt.strftime('%Y')
song_data['Month'] = song_data['date'].dt.strftime('%m')
song_data
```

## 2020 Billboard 

```python
year_2020 = song_data[song_data['Year']== '2020']
year_2020.shape
```

### 2020年有多少新歌？

```python
newentry = year_2020[year_2020['weeks-on-board']==1]
newentry
```

### 2020年每個月的新歌

```python
month_grouped = newentry.groupby('Month').sum()
month_grouped = month_grouped[['weeks-on-board']]
month_grouped.sort_values('weeks-on-board', ascending = False)
```

```python
month_grouped['weeks-on-board'].plot(color='red', figsize=(8,6))
plt.title('2020 new song')
plt.show()
```

### 2020年有哪些新歌進榜就拿到第一名

```python
newtopsong = year_2020[(year_2020['rank']==1) & (year_2020['weeks-on-board']==1)]
newtopsong
```

## 前10首在Billboard停留時間最久的歌

```python
song = song_data.groupby(['song','artist'])
```

```python
long = song_data[['song','artist']].value_counts()
long = long.reset_index(name='max') # reset index name
long = long[:10]
long 
```

```python
sns.barplot(data = long, x = 'max', y = 'song', palette = 'GnBu_d', alpha = 0.8)
plt.yticks(fontsize = 15)
plt.title('The songs appeared on board for the longest times', fontsize = 18)
for x,y in long["max"].items(): #To show the exact revenue generated on the figure
    plt.text(y,x,str(y), fontsize=12,color='k', verticalalignment='center')
plt.show()
```

## 獲得最多次第一名的歌手

```python
top_artists = song_data[song_data['rank']==1] # exract the artists who ranked first place
top_artists = top_artists.groupby('artist').sum() # sum of the rank
top_artists[['rank']].sort_values('rank', ascending = False).head(10)
```

```python
topartists_rank = top_artists.sort_values('rank', ascending = False)['rank']
topartists_rank = topartists_rank[:10,] # for the first 10 artists
topartists_rank = topartists_rank.reset_index(name='rank')
topartists_rank
```

```python
plt.figure(figsize = (15,5)) # width and height of figure is defined in inches
plt.title("The top artist who took the most number of first place", fontsize=18)
plt.bar(topartists_rank["artist"], topartists_rank["rank"],color= '#BEC23F', linewidth = 1)
plt.xlabel("artist",fontsize=15) # x axis shows the customers 
plt.ylabel("rank",fontsize=15) # y axis shows the Revenue
plt.xticks(fontsize=12, rotation=25)
plt.yticks(fontsize=12)
```

## 上榜最多次的藝人

```python
df_artist = song_data['artist'].value_counts()
df_artist.head()
```

```python
df_artist = df_artist[:10,] # for the first ten artists
plt.figure(figsize=(15,8)) # figsize: inch by inch
sns.barplot(x = df_artist.index, y = df_artist.values, alpha=0.9, palette = 'hot')
plt.ylabel('Count', fontsize=18)
plt.xlabel('Artists name', fontsize=18)
plt.title('The artists appeared on the chart most frequently', fontsize=25)
plt.xticks(rotation=45, fontsize=15)
for x,y in enumerate(df_artist.values):plt.text(x,y,'%s'%y,ha='center')
plt.show()
```
